When we do code in pytorch we must define this tensor for  implementation . or convert it to tensor for more details

```python
# input
x = torch.tensor([2.0])

# weight and bias implementation
w = torch.tensor([0.5],requires_grad=True)
b = torch.tensor([1.0],requires_grad=True)

# neuron output

y=w*x+b

print("Input :",x)
print('weights :',w)
print("Bias", b)
print("Output", y)
```

# 2 layer  neural network formation in details

```python
layer1= nn.Linear(2,3)
layer2= nn.Linear(3,1)

# INPUTS
x=torch.tensor([1.0,2.0])

h=layer1(x)

y=layer2(h)

print("Inputs",x)

print('\n layer 1 outputs \n',h)
print('\n layeer 2 outputs \n',y)
```

```python
print("Layer1 Weights")
print(layer1.weight)

print("Layer1 Bias")
print(layer1.bias)

print("Layer1 Output")
print(h)

print("----------------")

print("Layer2 Weights")
print(layer2.weight)

print("Layer2 Bias")
print(layer2.bias)

print("Final Output")
print(y)
```

#looss calculation in this outputs we see that only one outputs and now we calculate  the overall loss off this

```python
target = torch.tensor([10.0])

loss = (y-target)**2

print(loss)
```

```python
# compute this gradients
loss.backward()
```

```python
print(layer1.weight.grad)
print(layer1.bias.grad)

print(layer2.weight.grad)
print(layer2.bias.grad)
```

# weights and bias upgrade formula in detials

```python
lr = 0.01

with torch.no_grad():

    layer1.weight -= lr * layer1.weight.grad
    layer1.bias -= lr * layer1.bias.grad

    layer2.weight -= lr * layer2.weight.grad
    layer2.bias -= lr * layer2.bias.grad
```

```python
layer1.weight.grad.zero_()
layer1.bias.grad.zero_()

layer2.weight.grad.zero_()
layer2.bias.grad.zero_()
```

**Full training process in pytorch**


```python
import torch
import torch.nn as nn
x = torch.tensor([
    [1.0],
    [2.0],
    [3.0],
    [4.0]

])

y = torch.tensor([
    [2.0],
    [3.0],
    [5.0],
    [6.0]
])
```

```python
model = nn.Linear(1,1)# here one input and one  outputs
criterion = nn.MSELoss()# define this loss function in details
optimizer = torch.optim.SGD(# optimizer define in this code
    model.parameters(),
    lr = 0.01
)
```

```python
# Trainning  loop in details with code explanation

for epoch  in range(100):
  prediction =  model(x)
  loss = criterion(prediction,y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 10 == 0:
    print(epoch,loss.item())
```

# sIMPLE PROJECT USING SIMPLE PYTORCH FORMULA IN DETAILS

```python
import  torch
import  torch.nn as nn
import torch.optim as optim

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
housing = fetch_california_housing()

x = housing.data
y = housing.target

print("Features shape ", x.shape)
print("Target shape : ", y.shape)
```

```python
print(housing.feature_names)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

so we code this  in pytorch for that reason we convert it in tensor

```python
x_train = torch.tensor(x_train,dtype=torch.float32)
x_test = torch.tensor(x_test,dtype = torch.float32)
y_train = torch.tensor(y_train,dtype=torch.float32)
y_test = torch.tensor(y_test,dtype=torch.float32)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
# declaire the overall model in details
model = nn.Sequential(
    nn.Linear(8,64),
    nn.ReLU(),

    nn.Linear(64,32),
    nn.ReLU(),

    nn.Linear(32,1)
)

print(model)
```

```python
# define this neural network using class
def __init__(self):
  super().__init__()
  self.network = nn.Sequential(
      nn.Linear(8,64),
      nn.ReLU(),

      nn.Linear(64,32),
      nn.ReLU(),

      nn.Linear(32,1)
  )

def forward(self,x):
  return self.network(x)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)
```

```python
epochs =100

for epoch in range(epochs):
  model.train()
  prediction = model(x_train)
  loss = criterion(prediction,y_train)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 10 ==0:
    print(
        f"Epoch {epoch :3d}|"
        f"Loss : {loss.item():4f}"
    )
```

```python

model.eval()
with torch.no_grad():

    test_predictions = model(x_test)

    test_loss = criterion(
        test_predictions,
        y_test
    )

print("Test MSE:", test_loss.item())
```

```python
rmse = torch.sqrt(test_loss)

print("RMSE:", rmse.item())
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(
    y_test.numpy(),
    test_predictions.numpy()
)

print("R²:", r2)
```
